```python
import os
import sys
import pathlib
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np

random.seed(42)
```

```python
def bgr2hsv(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

def bgr2ycbcr(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

def ViewColorSpaces(rgb_image):
    hsvim = bgr2hsv(rgb_image);
    yuvim = bgr2ycbcr(rgb_image);

    im = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(12, 12))

    plt.subplot(4,3,1);plt.imshow(im);plt.title('RGB Space\n\noriginal')
    plt.subplot(4,3,4);plt.imshow(im[..., 0], cmap='Reds');plt.title('\nred')
    plt.subplot(4,3,7);plt.imshow(im[..., 1], cmap='Greens');plt.title('\ngreen')
    plt.subplot(4,3,10);plt.imshow(im[..., 2], cmap='Blues');plt.title('\nblue')


    plt.subplot(4,3,2);plt.imshow(hsvim);plt.title('HSV Space\n\ncombination')
    plt.subplot(4,3,5);plt.imshow(hsvim[..., 0], cmap='gray');plt.title('\nhue')
    plt.subplot(4,3,8);plt.imshow(hsvim[..., 1], cmap='gray');plt.title('\nsaturation')
    plt.subplot(4,3,11);plt.imshow(hsvim[..., 2], cmap='gray');plt.title('\nbrightness')

    plt.subplot(4,3,3);plt.imshow(yuvim);plt.title('YCbCr Space\n\ncombination')
    plt.subplot(4,3,6);plt.imshow(yuvim[..., 0], cmap='gray');plt.title('\nLuminance')
    plt.subplot(4,3,9);plt.imshow(yuvim[..., 1], cmap='gray');plt.title('\nblue difference')
    plt.subplot(4,3,12);plt.imshow(yuvim[..., 2], cmap='gray');plt.title('\nred difference')

    plt.show()
```

```python
train_root = pathlib.Path('../../FruitScale/dataextended/Training')
train_root = train_root.resolve()
```

```python
x=random.choice(os.listdir(str(train_root)+'/Banana'))
img=cv2.imread(os.path.join(str(train_root)+'/Banana',x))

ViewColorSpaces(img)
```
